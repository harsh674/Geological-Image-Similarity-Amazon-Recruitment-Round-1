# Dealing with Deep Learning

```python
dl_data=np.concatenate((schist_data, rhyolite_data,quartzite_data,marble_data,gneiss_data,andesite_data), axis=0)
```

```python
dl_data = np.array(dl_data)
```

### Labeling Data 1-schist , 2- rhyolite , 3-quartzite , 4-marble , 5-gneiss , 6-andesite

```python
label_data = []
```

```python
for i in range(5000):
    label_data.append(1)
for i in range(5000):
    label_data.append(2)
for i in range(5000):
    label_data.append(3)
for i in range(4998):
    label_data.append(4)
for i in range(5000):
    label_data.append(5)
for i in range(5000):
    label_data.append(6)
    
```

```python
len(label_data)
```

```python
from tensorflow import keras
```

```python
import random
c = list(zip(dl_data, label_data))

random.shuffle(c)

dl_data, label_data = zip(*c)
X_tr,X_te,Y_tr,Y_te = train_test_split(dl_data,label_data , test_size=0.33)
```

```python
GeographicalModel = keras.Sequential([
                               keras.layers.Flatten(input_shape = (28,28,3)), # as we have seen our data is in 28,28,3
                               keras.layers.Dense(128,activation='relu'),
                               keras.layers.Dense(10,activation = 'softmax')
])
GeographicalModel.compile(optimizer = 'adam' , loss = 'sparse_categorical_crossentropy' , metrics = ['accuracy'])
GeographicalModel.fit(x=np.array(X_tr), y=np.array(Y_tr),epochs = 10)
```

```python
y_pred = []
for i in range(len(X_te)):
    img = X_te[i]
    img = (np.expand_dims(img,0)) 
    result = GeographicalModel.predict(img) 
    correctResult = np.argmax(result)
    y_pred.append(correctResult)
```

```python
y_pred
```

```python
accuracy_score(Y_te, y_pred)
```
